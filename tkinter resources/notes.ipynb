## Unlimited Arguments


```python
def add(*args):
    sum=0
    for n in args:
        sum+=n

    return sum

print(add(5,2,3,4))
```

# **Kwargs: Many Keywrorded Arguments

```python
def calculate(n, **kwargs):
    #print(kwargs)
    n+=kwargs['add']
    n*=kwargs['multiply']
    return n
print(calculate(2, add=3, multiply=5))
```
