### Tag: 
A Tag object corresponds to an XML or HTML tag in the original document.



```python
soup = BeautifulSoup('<b class="boldest">Extremely bold</b>', 'html.parser')
tag = soup.b
type(tag)
# <class 'bs4.element.Tag'>
```

##### Name
Every tag has a name:

```python
tag.name
```

If you change a tag’s name, the change will be reflected in any markup generated by Beautiful Soup down the line:

```python
soup = BeautifulSoup(content_html, 'html.parser')
anchor_tag= soup.a 
anchor_tag.name = "link"
soup.prettify
```

### attrs¶
An HTML or XML tag may have any number of attributes. The tag <b id="boldest"> has an attribute “id” whose value is “boldest”. You can access a tag’s attributes by treating the tag like a dictionary:




```python
tag = BeautifulSoup('<b id="boldest">bold</b>', 'html.parser').b
tag['id']
# 'boldest'
```

```python
tag.attrs
```

You can add, remove, and modify a tag’s attributes. Again, this is done by treating the tag as a dictionary:

```python
tag['id'] = 'verybold'
tag['another-attribute'] = 1
tag
```

```python
del tag['id']
del tag['another-attribute']
tag
```

```python
# tag['id']
# # KeyError: 'id'

tag.get('id')
# None
```

#### Multi-valued attributes
<p>HTML 4 defines a few attributes that can have multiple values. HTML 5 removes a couple of them, but defines a few more. The most common multi-valued attribute is class (that is, a tag can have more than one CSS class). Others include rel, rev, accept-charset, headers, and accesskey. By default, Beautiful Soup parses the value(s) of a multi-valued attribute into a list:
</p>

```python
css_soup = BeautifulSoup('<p class="body"></p>', 'html.parser')
css_soup.p['class']
```

```python
css_soup = BeautifulSoup('<p class="body strikeout"></p>', 'html.parser')
css_soup.p['class']
# ['body', 'strikeout']
```

If an attribute looks like it has more than one value, but it’s not a multi-valued attribute as defined by any version of the HTML standard, Beautiful Soup will leave the attribute alone:

```python
id_soup = BeautifulSoup('<p id="my id"></p>', 'html.parser')
id_soup.p['id']
# 'my id'
```

When you turn a tag back into a string, multiple attribute values are consolidated:



```python
rel_soup = BeautifulSoup('<p>Back to the <a rel="index first">homepage</a></p>', 'html.parser')
rel_soup.a['rel']
# ['index', 'first']
rel_soup.a['rel'] = ['index', 'contents']
print(rel_soup.p)
# <p>Back to the <a rel="index contents">homepage</a></p>
```

You can force all attributes to be parsed as strings by passing multi_valued_attributes=None as a keyword argument into the BeautifulSoup constructor:

```python
no_list_soup = BeautifulSoup('<p class="body strikeout"></p>', 'html.parser', multi_valued_attributes=None)
no_list_soup.p['class']
# 'body strikeout'
```

## class bs4.NavigableString


A string corresponds to a bit of text within a tag. Beautiful Soup uses the NavigableString class to contain these bits of text:



```python
soup = BeautifulSoup('<b class="boldest">Extremely bold</b>', 'html.parser')
tag = soup.b
tag.string
# 'Extremely bold'
type(tag.string)
```

A NavigableString is just like a Python Unicode string, except that it also supports some of the features described in Navigating the tree and Searching the tree. You can convert a NavigableString to a Unicode string with str:

```python
unicode_string = str(tag.string)
unicode_string
# 'Extremely bold'
type(unicode_string)
# <type 'str'>
```

You can’t edit a string in place, but you can replace one string with another, using replace_with():

```python
tag.string.replace_with("No longer bold")
tag
# <b class="boldest">No longer bold</b>
```
